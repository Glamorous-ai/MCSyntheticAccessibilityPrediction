## This notebook converts SMILES strings into Morgan Fingerprints

```python
import pandas as pd
import timeit
import scipy.sparse as sp

df_train = pd.read_csv("data/df_87w_five_classes.csv")
print(df_train.shape)
```

```python
df_train.smiles.tolist()
```

```python
from tdc.chem_utils import MolConvert

NUM_2D_FEATURES=['ECFP2','ECFP4','ECFP6','MACCS','Daylight','RDKit2D','Morgan','PubChem']

def get_features(smiles,target_features):
    if target_features==None or target_features=='':
        return smiles
    converter = MolConvert(src = 'SMILES', dst = target_features)
    features=converter(smiles)
    return features
```

```python
start = timeit.default_timer()
sparse_train = sp.csr_matrix(get_features(df_train.smiles.tolist(),'Morgan'))
stop = timeit.default_timer()
print('Time: ', stop - start)  

sp.save_npz('feature/87w_morgan2.npz', sparse_train)
```
